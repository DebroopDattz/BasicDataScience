```python
import pandas as pd
import numpy as np
import matplotlib . pyplot as plt
from sklearn . linear_model import LinearRegression
from sklearn . metrics import mean_squared_error , r2_score
```

```python
data = pd. DataFrame ({
'Hours': [1, 2, 3, 4.5 , 5, 6, 7, 8.5 , 9, 10] ,
'Marks': [35 , 40, 50, 60, 62, 70, 75, 85, 88, 95]
})
```

```python
X = data[['Hours']]
y = data ['Marks']
```

```python
model = LinearRegression()
model.fit(X,y)

y_pred = model.predict(X)
print(f"Intercept: {model.intercept_: .2f}")
print(f"Slope: {model.coef_[0]: .2f}")
mse = mean_squared_error(y, y_pred)
r2= r2_score(y, y_pred)
print (f"Mean Squared Error : {mse :.2f}")
print("R_squared score: ",r2)

plt.figure(figsize=(8,5))
plt. scatter (X, y, color ='blue', label ='Actual Marks')
plt.plot(X, y_pred, color ='red', linewidth=2, label ='Regression Line')

plt.title('Hours vs Marks')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')    
plt.legend()
plt.grid(True)
plt.show()
```

```python
test_hours = [[6.5]]
predicted_marks = model.predict(test_hours)
print(f"Predicted marks for 6.5 study hours: {predicted_marks[0]: .2f}")
```
