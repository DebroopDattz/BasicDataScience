```python
#1.Write a Python program to create a plot to get a general Statistics of Iris data. (Using dataset iris.csv)
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
#data=sns.load_dataset("iris")
data = pd.read_csv('iris.csv')
data.describe().plot(kind="area",fontsize=27,figsize=(20,4),table=True)
plt.xlabel("stat")
plt.ylabel("val")
plt.title("general stat of iris dataset")
#print(data.describe()) # Display general statistics of the dataset
```

```python
#3.Write a Python program to create a join plot using "hexbin" to describe individual
#distributions on the same plot between Sepal length and Sepal width. (Using iris.csv).
import pandas as pd
import seaborn as sns
data = pd.read_csv('iris.csv')
fig=sns.jointplot(x='SepalLengthCm', y='SepalWidthCm',kind="hex",color="purple",data=data)
```

```python
#4.Write a Python program to create a join plot and add regression and kernel density fits using "reg" to describe 
#individual distributions on the same plot between Sepal length and Sepal width. (Using the dataset iris.csv).
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('iris.csv')

# Create a joint plot with regression and kernel density estimation
fig = sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', kind='reg', color='blue', data=data,
                    joint_kws={'scatter_kws': {'alpha': 0.4}, 'line_kws': {'color': 'red'}})

# Add kernel density estimation (KDE) plots to the sides of the joint plot
sns.kdeplot(data['SepalLengthCm'], ax=fig.ax_marg_x, color='blue')
sns.kdeplot(data['SepalWidthCm'], ax=fig.ax_marg_y, color='blue', vertical=True)
fig=sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',kind='reg',color='blue',data=data)
plt.show()
```
