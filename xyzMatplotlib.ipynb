```python
#1. Take a data set in a data frame and plot the bar chart example may be year of Production
#   and the Production Quantity. (Take help of Pandas)

import matplotlib.pyplot as plt
import pandas as pd
dict={"Year of production":[2011,2015,2019,2023],"Production Quantity":[830,501,1800,400]}
df=pd.DataFrame(dict)
print(df)
x=df['Year of production']
y=df['Production Quantity']
plt.bar(x,y,width=2)
plt.xlabel('Year of production')
plt.ylabel('Production Quantity')
plt.title('B A R    C H A R T')
plt.show()
```

```python
#2. Take a data set from the above data frame which you have prepared and plot the pie
#   chart. (Take help of Pandas)

import matplotlib.pyplot as plt
import pandas as pd
Year_of_production=[2011,2015,2019,2023]
Production_Quantity=[830,501,1800,400]
dict={"Year of production":[2011,2015,2019,2023],"Production Quantity":[830,501,1800,400]}
df=pd.DataFrame(dict)
print(df)
fig1,ax1=plt.subplots()
ax1.pie(Production_Quantity,labels=Year_of_production,autopct='%1.1f%%',shadow=False,startangle=90)
ax1.axis('equal')
plt.title('P I E   C H A R T')
plt.show()
```

```python
#3. Take a csv file with name, employee id and salary. Read the data of salary and plot the bar
#   Chart with employee name on the x axis and employee salary on the y-axis. Just take 5
#   records in the csv files.

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Position_Salaries - Position_Salaries.csv')
df1=df.head(5)
print(df1)
x=df1['Name']
y=df1['Salary']
plt.bar(x,y)
plt.xlabel('Name->')
plt.ylabel("Salary->")
plt.title("Bar Chart with employee name and salary")
plt.show()
```

```python
#4. Take the dirtydata.csv as an input and plot the box-plot for the column duration and find
#   out. How many outliers are there?

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('dirtydata.csv - dirtydata.csv.csv')
x=df['Duration']
plt.boxplot(x)
plt.show()
#outliers->
Q1=x.quantile(0.25)
Q3=x.quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers=df[(x<lower_bound)|(x>upper_bound)]
print("Outliers are\n ",outliers)
print("no of outliers =",len(outliers))
```

```python
#5.What do you mean by multiple box plots? Take a data set in a data frame and draw the
#multiple box plots. Make a csv file with multiple parameters and then plot multiple box plots
#for each parameter.

import pandas as pd
import matplotlib.pyplot as plt
dict={"Year of production":[2011,2015,2019,2023],"Production Quantity":[830,501,1800,400]}
df1=pd.DataFrame(dict)
print(df1)
plt.figure(figsize=(4,2))
df1.boxplot(column=["Year of production","Production Quantity"])
plt.title("Multiple box plots for data set in a data frame---->")
plt.show()
df=pd.read_csv('dirtydata.csv - dirtydata.csv.csv')
plt.figure(figsize=(10,6))
df.boxplot(column=['Duration','Pulse','Maxpulse','Calories'],grid=True)
plt.title("Multiple box plots for CSV file")
plt.show()
```
