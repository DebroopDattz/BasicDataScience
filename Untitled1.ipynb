```python
import matplotlib.pyplot as plt
import numpy as np

X= np.array(['Deb', 'Abhi', 'Oli', 'Megh'])
Y=np.array([5.9,4.7,3.7,2.6])
plt.bar(X,Y)
plt.title("Height Chart")
plt.xlabel("Name of Student")
plt.ylabel("Height")
plt.bar(X,Y,color="red")
plt.bar(X,Y,width=0.9)
plt.show()
```

```python
import pandas as pd
data={'Year of Production':['2021','2022','2023','2024'],'Units of Production':[2000,3000,3500,2500]}
df=pd.DataFrame(data)
X=df['Year of Production']
Y=df['Units of Production']
plt.bar(X,Y)
plt.title("Production Chart")
plt.xlabel("Y.O.P")
plt.ylabel("U.O.P")
plt.show()
plt.pie(Y,labels=X,autopct='%1.1f%%')
plt.title("Production Pie Chart")
plt.xlabel('Y.O.P')
plt.ylabel('U.O.P')
#plt.pie()
plt.show()
```

```python
data={'Employee Name':['ABC','DEF','TRV','XYZ'],'Salary':[25000,70000,125000,35000]}
df=pd.DataFrame(data)
df.head()
X=df['Employee Name']
Y=df['Salary']
plt.title("Employee Name vs Salary")
plt.xlabel("Name of Employee")
plt.ylabel("Salary of Employee")
plt.bar(X,Y)
plt.show()
```
